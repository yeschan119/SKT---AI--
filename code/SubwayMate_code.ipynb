```python
#********* 이 표시는 나중에 parameter value로 내보낼 값들
#사용 하는 모듈은 아래와 같다.
#reuests : 파이썬에서 HTTP에 요청을 보냄
# json : json파일을 다루기 위한 모듈
#urllib은 http에 request를 요청하여 페이지를 읽어오는 모듈이다.
#python 3.0이상은 urllib.request import urlopen
#python 3.0아래는 urllib import urlopen을 사용한다.
# xml.etree.ElementTree : 요청한 파일이 XML형식일 경우 트리로 나타내어 하나씩 파씽해주는 API
import requests
import json
import time
from time import gmtime, strftime
import pandas as pd
import xml.etree.ElementTree as ET
import sys
if sys.version_info[0] == 3:
    from urllib.request import urlopen
else:
    from urllib import urlopen
from bs4 import BeautifulSoup

#machine learning
from sklearn.model_selection import train_test_split # train, test 데이터 분할
from sklearn.cluster import KMeans #비지도 학습 모델 중 k-menas모듈
from sklearn.linear_model import LinearRegression #선형 회귀분석
import matplotlib.pyplot as plt   # plot 으로 시각화
import seaborn as sb  #시각화 모듈
from sklearn.preprocessing import LabelEncoder   #text를 숫자로 변환해주는 모듈
import datetime #현재 시간을 알기 위한 모듈
from IPython.core.display import display


def get_subway_info(departure, arrival):
    #지하철에서 지상까지 올라오는 시간을 체크하기 위해 에스컬레이터 길이를 담은 데이터 불러오기
    escalator = pd.read_excel('/Users/Eungchan/desktop/Ai_project/escalator.xlsx', header = 2, encoding='utf-8')
    escalator_index = escalator.index
    escalator_speed = 30  # 한국지하철 표준 속도 30m/min
    
    # 로컬저장소에 있는 csv파일을 불러서 subway_code에 대한 정보를 얻는다.
    subway_code = pd.read_csv('/Users/Eungchan/desktop/Ai_project/subway_code.csv')
    
    # odsay Lab(대중교통 open API)홈페이지를 이용해서 지하철 정보(소요시간, 경유역, 환승정보, 플랫폼 등)를 가져온다.
    # 발급받은 api key를 이용해 URL을 string type으로 만든다.
    # 출발역코드와 도착역코드를 기준으로 데이터를 불러오기 때문에 그 부분을 비워 놓고 *로 채워 놓는다.(split을 위해서)
    # 함수인자로 출발역코드와 도착역코드를 받으면 별표자리에 그 값들을 넣고 request를 요청하여 지하철 정보를 불러온다.
    def get_subwayObject(start, stop):
        url = 'https://api.odsay.com/v1/api/subwayPath?lang=0&CID=1000&SID=*&EID=*&Sopt=1&apiKey=hOy%2F5bdCEg8YAM%2FW1YkgDKz9AF%2FGuvkKwdva8pid%2FuA'
        list_URL = url.split('*')
        URL = list_URL[0] + start + list_URL[1] + stop + list_URL[2]
        response = requests.get(URL)
        response.status_code
        text = response.text
        jasonObject = json.loads(text)
        return jasonObject

    #출발역과 도착역정보는 역코드로 입력된다.
    #사용자에게서 입력되는 정보는 역 이름이므로 역 이름을 가지고 역 코드를 리턴해주는 함수
    def get_station_code(start, stop):
        start = ''
        stop = ''
        index = get_index()  # get index for json data
        for i in range(index):
            if subway_code['전철역명'][i] == departure and start == '':
                start = subway_code['외부코드'][i]
            elif subway_code['전철역명'][i] == arrival and stop == '':
                stop = subway_code['외부코드'][i]
        return start, stop
    #출발역과 도착역사이에 몇 개의 역을 지나야 하는지 알려주는 함수
    #json형식의 데이터를 받아서 경유역을 모두 리스트에 담아 리턴    
    def get_stationCount(Object):
        stations = []
        for station in Object['result']['stationSet']['stations']:
            stations.append(station.get('startName'))
        return stations
    #지하에서 지상까지 올라가는 시간을 체크
    #에스컬레이터 길이를 얻어
    #국내 표준 에스컬레이터 속도(escalator_speed)를 이용하여 시간 계산
    def get_escalator_time(stop):
        check_escalator = 0
        for i in escalator_index:
            if escalator['역명'][i] == stop:
                check_escalator = 1
                return(int((escalator['승강행정\n(m)'][i]/escalator_speed) * 100))
        if(check_escalator == 1):
            return 0
        
    #지허철 정보(string or list or dataFile, json)가 어떤 형식이든 인덱싱을 하면 값을 추출하기 편하다
    # data.index와 같이 indexing이 가능한 형식도 있지만 int형이 아니고 모든 데이터형식에 맞는 것도 아니라서 함수 사용
    # get_index()함수는 각 데이터를 인덱싱하여 int값으로 리턴해준다.
    def get_index():
        counter = 0;
        index = subway_code.index
        for i in index:
            counter += 1
        return counter

    #환승이 필요할 경우 환승이 필요한 역만 따로 뽑아서
    #정확히 어디서 환승하고 얼마나 걸어야 하고 어느 플랫폼이 가장 빨리 갈 수 있는지를 알려주는 함수
    #환승정보는 하나의 string으로 넘어와서 거기서 필요한 정보만 뽑아야 한다.
    #string -> list -> dictionary로 파일을 변환하여 리턴
    #string -> list 과정에 불필요한 특수기호가 하나의 문장으로 붙어있어서 아래와 같이 불필요한 제거작업 진행
    def get_exchange_info(Object):
        dic = {}
        ExName = Object["result"]["exChangeInfoSet"]['exChangeInfo']
        Ex_str = str(ExName)
        Ex_str = Ex_str.replace(' ','')
        Ex_str = Ex_str.replace(',',':')
        Ex_str = Ex_str.replace('[','')
        Ex_str = Ex_str.replace(']','')
        Ex_str = Ex_str.replace('{','')
        Ex_str = Ex_str.replace('}','')
        Ex_str = Ex_str.replace("'",'')
        Ex_list = Ex_str.split(':')

        #list를 key : value형식으로 넘겨주기 위해 key와 value를 2씩 증가시키면서 dic에 저장
        # ex key = list[0] : value = list[1] 
        key = 0
        value = 1
        for value in range(1, len(Ex_list),2):
            dic[Ex_list[key]] = Ex_list[value]
            key += 2
            value += 2
        print(dic['exName']+"역에서 내려",dic['exWalkTime'],"미터 걸어가야 합니다.")
        print("빠른 환승은",dic['fastDoor']+"번입니다.")

    #데이터를 처리할 때마다 에러체크를 해주어 갑자기 멈추는 현상 방지
    def error_check(Object, value):
        for check_trans in Object["result"]:
            if check_trans == value:
                return True
        return False

    #main function
    start, stop = get_station_code(departure, arrival)  #get station code
    jasonObject = get_subwayObject(start, stop)  # get subway information using code
    print()
    time.sleep(1)
    if error_check(jasonObject, 'globalTravelTime'):
        print("역간 소요시간은",jasonObject["result"]['globalTravelTime'],"분")
    else:
        print("찾으시는 역정보가 없습니다.")
    print()
    time.sleep(1)
    print("소요거리는",jasonObject["result"]['globalDistance'],"km")
    print()
    time.sleep(1)
    
    #환승이 필요한 경우만 걸러내기 위해 if 사용
    if error_check(jasonObject, "exChangeInfoSet"):
        get_exchange_info(jasonObject)
    else:
        print("환승정보가 없습니다.")
    print()
    time.sleep(1)
    global global_station
    global_station = get_stationCount(jasonObject) #get all required stations
    for station in global_station:
        print(station,end=" -> ")
    if error_check(jasonObject,'globalStationCount'):
        print(arrival+"까지 모두",jasonObject["result"]['globalStationCount'],"개 역입니다.")
    else:
        print("경유역 정보가 존재하지 않습니다.")
    
    print()
    time.sleep(1)
    escalator_time = get_escalator_time(arrival)   #에스컬레이터 시간 얻기
    #if escalator_time != 0:
        #print(arrival,"역에서 지상까지 올라가는 시간은",(escalator_time//60),'분', (escalator_time % 60),'초 걸립니다.')
    #else:
       # print(arrival,"역은 에스컬레이터가 존재하지 않습니다.")
        
    
def get_realTime_arrival(arrival_station):            
    #지하철 정보를 실시간으로 보내주는 URL을 이용해서 데이터를 불러온다.
    # get realtime_station_arrival API
    URL = 'http://swopenapi.seoul.go.kr/api/subway/5862464c5079657335396377764572/xml/realtimeStationArrival/ALL/1/5/'
    response = urlopen(URL).read()
    # parsing the xml
    xtree = ET.fromstring(response)

    #필요한 값들을 리스트에 담기
    rows = []

    #xml파일을 파싱하여 필요한 데이터만 추출하여 필요한 리스트 생성
    # extract field using parsing
    #node.find()안에 들어 있는 데이터가 원하는 목록
    for node in xtree.findall("row"):
        n_subwayId = node.find("subwayId").text
        n_trainLineNm = node.find("trainLineNm").text
        n_subwayHeading = node.find("subwayHeading").text
        n_statnNm = node.find("statnNm").text
        n_recptnDt = node.find("recptnDt").text    
        n_arvlMsg2 = node.find("arvlMsg2").text
        rows.append({"역아이디": n_subwayId, 
                     "운행노선" : n_trainLineNm,
                     "도착방면": n_subwayHeading, 
                     "역이름": n_statnNm,
                     "도착시간": n_recptnDt[10:18],  # only get time, not year, date
                     "도착메시지" : n_arvlMsg2})

    
    #리스트로 작성한 값을 DataFrame으로 변환
    #리스트에 column을 넣어서 dataFrame 생성
    columns = ["역아이디", "운행노선", "도착방면", "역이름","도착시간","도착메시지"]
    real_time_subway = pd.DataFrame(rows, columns = columns)
    #print("example : ")
    #print(real_time_subway.head(10))
    print()
    print("===============================\n")

    #검색어랑 일치하는 역을 찾아서 실시간 데이터를 리스트로 출력
    # real_time_subway.iloc[index].tolist() 리스트 전환코드
    #*********
    def output_realTimeinfo(real_time_subway):
        index = 0
        for station_name in real_time_subway['역이름']:
            if station_name == arrival_station:
                df_to_list = real_time_subway.iloc[index].tolist()
                print(real_time_subway.iloc[index])
                time.sleep(1)
                print("\n")
            index += 1
        return index
    print()
    print(arrival_station+"역의 실시간도착 정보입니다.")
    print()
    #리턴값이 0이면 찾는 역이 없는 것
    if output_realTimeinfo(real_time_subway) == 0:
        print("찾는 역이 없습니다.")
        
#predict congestions in subway using past data
def execute_congestion_check(X):
    #지하철 혼잡도 기준
    congesstion_rate = 0.5
    #그래프 배경화면을 다크하게
    #그래프 배경화면을 다크하게
    #지하철 데이터를 불러오는 함수
    plt.style.use(['dark_background'])
    congestion = pd.read_excel('/Users/Eungchan/desktop/Ai_project/2019congestion.xls')
    def get_week_weekend(congestion):
        week_index = 0
        saturday_index = 0
        sunday_index = 0
        for day in congestion['조사일자']:
            if day == '평일':
                week_index += 1
            elif day == '토요일':
                saturday_index += 1
            else:
                sunday_index += 1
        saturday_index += week_index
        #평일
        week = congestion.head(week_index)
         #토요일
        saturday = congestion[week_index : saturday_index]
        #일요일
        sunday = congestion.tail(sunday_index)
        return week, saturday, sunday

    def get_weekdays_line(weekdays):
        week_line = []
        temp_line = '1호선'
        index = 0
        start = 0
        for line in weekdays['호선']:
            if line == temp_line:
                index += 1
            else:
                index += 1
                week_line.append(weekdays[start:index-1])
                start = index
                temp_line = line
        week_line.append(weekdays[start:index-1])
        return week_line;

    def drop_duplicate(week):
        for i in range(len(week)):
            if week['구분1'][i] == '상선' or week['구분1'][i] == '외선':
                week.drop(i, inplace = True)
        return week

    def get_weekend_line(weekend):
        weekend_line = []
        temp_line = '1호선'
        index = 0
        start = 0
        for line in weekend['호선']:
            if line == temp_line:
                index += 1
            else:
                index += 1
                weekend_line.append(weekend[start:index-1])
                start = index
                temp_line = line
        weekend_line.append(weekend[start:index-1])
        return weekend_line

        #현재시간을 얻어서 시간과 분만 리턴
    def get_current_time():
        current_time = datetime.datetime.now()
        if int(str(current_time.strftime('%M'))) >= 30:
            current_time = datetime.time(int(str(current_time.strftime('%H'))),30)
        else:
            current_time = datetime.time(int(str(current_time.strftime('%H'))),0)
        return current_time

    def get_X_Y(X, line_data):
        Y = []
        high_congest = []
        for i in range(len(line_data)):
            for station in X:
                if line_data['역명'][i:i+1].values == station:
                    Y.append(line_data[current_time][i:i+1].values)
                    if line_data['기준'][i:i+1].values == 1:
                        high_congest.append(station)
        return X, Y, high_congest

    def add_column(current_time, line_data):
        standard = []
        for conges_rate in line_data[current_time]:
            if conges_rate > 0.3:
                standard.append(1)
            else:
                standard.append(0)
        line_data['기준'] = standard
        return line_data

    def build_graph_with_congestion(X,Y,line_data, high_congestion):
        #그래프 사이즈 조정(pixel단위로)
        fig, ax = plt.subplots()
        x = 700 / fig.dpi
        y = 250 / fig.dpi
        fig.set_figwidth(x)
        fig.set_figheight(y)
        #그래프 그리기
        time = datetime.datetime.now().strftime("%H:%M")
        title = str(time)+'분' + " "+line_data['호선'][0:1].values +' 혼잡도'
        new_X = []
        new_Y = []
        index = 0
        for x,y in zip(X,Y):
            markercolor = 'blue'
            color = 'blue'
            new_X.append(x)
            new_Y.append(y)
            index += 1
            if index == 2:
                if new_X[0] in high_congestion and new_X[1] in high_congestion:
                    markercolor = 'red'
                    color = 'red'
                plt.plot(new_X,new_Y,color=color,marker='o',linestyle='solid', markerfacecolor=markercolor,markersize=10)
                del new_X[0]
                del new_Y[0]
                index = 1
        plt.title(title,loc='center')
        plt.xlabel('역이름')
        plt.ylabel('혼잡도')
        plt.show()  

    def build_graph(X,Y, line_data):
        fig, ax = plt.subplots()
        x = 700 / fig.dpi
        y = 250 / fig.dpi
        fig.set_figwidth(x)
        fig.set_figheight(y)
        #그래프 그리기
        time = datetime.datetime.now().strftime("%H:%M")
        title = str(time)+'분' + " "+line_data['호선'][0:1].values +' 혼잡도'
        plt.plot(X,Y,color='blue',marker='o',linestyle='solid', markerfacecolor='blue',markersize=11)

        plt.title(title,loc='center')
        plt.xlabel('역이름')
        plt.ylabel('혼잡도')
        plt.show()

        #cluster 시각화
    def build_clusters(XX,YY):
        conges_degree = pd.DataFrame(columns=('역번호', '혼잡도'))
        conges_degree['역번호'] = XX[0].values
        conges_degree['혼잡도'] = YY.values
        k_views = KMeans(n_clusters=3).fit(conges_degree)
        conges_degree['cluster'] = k_views.labels_
        sb.lmplot('역번호','혼잡도', data=conges_degree, fit_reg=False, scatter_kws={"s":100}, hue='cluster')
    def check_week():
        week = datetime.datetime.today().weekday()
        return week
    def get_line(station, lines):
        for i in range(len(lines)):
            j = i+1
            line = str(lines['역명'][i:j].values)
            line = line.replace('[','')
            line = line.replace(']','')
            line = line.replace("'",'')
            if line == station:
                temp = str(lines['호선'][i:j].values)
                return (int(temp[2]) - 1)
        return -1


    congestion = drop_duplicate(congestion)
    if get_line(X[0],congestion) >= 0:
        line_num = get_line(X[0],congestion)
    else:
        print("찾는 역이 없습니다.")

    weekdays, saturday, sunday = get_week_weekend(congestion)
    if check_week() < 5:
        week_line = get_weekdays_line(weekdays)
        line_data = week_line[line_num]
    elif check_week() == 5:
        saturday_line = get_weekend_line(saturday)
        line_data = saturday_line[line_num]
    else:
        sunday_line = get_weekend_line(sunday)
        line_data = sunday_line[line_num]

    current_time = get_current_time()
    #머신러닝
    line_data = add_column(current_time, line_data)
    # pyplot
    #X = ['왕십리','상왕십리','신당','동대문역사문화공원','을지로4가','을지로3가']
    X, Y, high_congestion = get_X_Y(X, line_data)

    if len(high_congestion) != 0:
        print(high_congestion,"<- 이 역들은 지금 시간에 혼잡합니다.")
    else:
        print("현재시간에는 혼잡한 구간이 없습니다.")

    #한글폰트 삽입 함수
    plt.rc('font', family='AppleGothic')
    #build_graph(X,Y, line_data)
    build_graph_with_congestion(X,Y,line_data,high_congestion)
    #clustering    
    XX = pd.DataFrame(line_data['역번호'].values)
    YY = pd.DataFrame(line_data[current_time].values)

    x_train, x_test, y_train, y_test = train_test_split(XX,YY, test_size=0.2,random_state=0)

    # 선형회귀분석
    clf = LinearRegression()  
    clf.fit(x_train, y_train)  #모수 추정
    clf.coef_    #추정된 모수 확인
    clf.intercept_  # 추정 된 상수항 확인
    clf.predict(x_test) # 예측
    clf.score(x_test, y_test) # 모형 성능 평가
    print("회귀분석 예측결과",clf.score(x_test, y_test))
    #clustering
    #비지도 학습 모델
    k_means = KMeans(n_clusters=2).fit(x_train)
    predict_cluster = k_means.predict(x_test)
    print("비지도학습 예측:",predict_cluster)

    build_clusters(XX,YY)



    
              
Exit =''  # while loop를 끝내기 위한 변수            
while True:
    departure = input("출발역 입력 : ")
    arrival = input("도착역 입력 : ")
    get_subway_info(departure, arrival)
    real_time = input("실시간 도착정보를 원하시나요?(y/n)")
    if real_time != 'n':
        get_realTime_arrival(departure)
    print()
    congestion_control = input("혼잡도를 보시겠어요?(y/n)")
    if congestion_control != 'n':
        execute_congestion_check(global_station)
    Exit = input("계속 하시겠습니까?(1/0) : ")
    if Exit == '0':
        break
```
